4 - Quadrado a cubo

Crie uma fun√ß√£o chamada *calcular_potencias* que recebe um n√∫mero *n* e retorne o quadrado e o cubo desse n√∫mero em uma tupla.

A fun√ß√£o deve retornar dois valores. Lembre que em Python, voc√ä pode retornar m√∫ltiplos valores, que s√£o automaticamente empacotados em uma tupla.

```python
def calcular_potencias (n):
  resultado_quadrado = n ** 2
  resultado_cubo = n ** 3
  return resultado_quadrado, resultado_cubo

numero = float (input("Insira um n√∫mero para calcul√°-lo ao cubo e ao quadrado: "))

print(f"O n√∫mero {numero} ao quadrado e ao cubo √©: ", calcular_potencias(numero))
```

5 - Contagem regressiva
Crie uma fun√ß√£o contagem_regressiva que recebe um n√∫mero inteiro inicio e imprime todos os n√∫meros de inicio at√© 1.

Use um la√ßo de repeti√ß√£o for ou while para iterar e imprimir os n√∫meros sequencialmente. N√£o √© necess√°rio retornar nada.

6 - C√°lculo de Desconto

Desenvolva uma fun√ß√£o chamada calcular_desconto que aceita valor_original e percentual_desconto como par√¢metros e retorna o valor final ap√≥s aplicar o desconto.

$$
\text{Valor Final} = \text{Valor Original} \times (1 - \frac{\text{Percentual Desconto}}{100})
$$

 Para calcular o percentual corretamente, divida o percentual_desconto por 100 antes de subtrair de 1.

```python
def calcular_desconto(valor_original, percentual_desconto):
  valor_final = valor_original * (1 - percentual_desconto/100)
  return valor_final

valor_input = float (input("Insira o valor orignal do produto: "))
desconto_input = float (input("Insira o percentual do desconto: "))

print ("O valor final √© de: ", calcular_desconto(valor_input, desconto_input))
```

7 - Maior e Menor

Crie uma fun√ß√£o chamada encontrar_extremos que recebe uma lista de n√∫meros e retorna o maior e o menor n√∫mero da lista.

 Voc√™ pode explorar as fun√ß√µes nativas do Python max() e min() para uma solu√ß√£o concisa, ou implementar a l√≥gica de compara√ß√£o manualmente.

```python
def encontrar_extremos (lista):
  if not lista:
        return None
  return min(lista), max(lista)


numeros = [5, 2, 9, 1, 7, 3]
resultado = encontrar_extremos(numeros)
print(f"Lista: {numeros}")
print(f"Menor: {resultado[0]}, Maior: {resultado[1]}")
```

8 - Calculadora Simples

Crie uma fun√ß√£o chamada calculadora que recebe dois n√∫meros (a e b) e uma operacao (string: 'soma', 'subtracao', 'multiplicacao', 'divisao') e retorna o resultado da opera√ß√£o.

 Use estruturas condicionais (if/elif/else) para determinar qual c√°lculo realizar e adicione uma verifica√ß√£o para evitar divis√£o por zero.


```python
def calculadora(a, b, operacao):

    operacao = operacao.lower().strip()

    if operacao == 'soma':
        return a + b
    elif operacao == 'subtracao':
        return a - b
    elif operacao == 'multiplicacao':
        return a * b
    elif operacao == 'divisao':
        if b == 0:
            return "Erro: Divis√£o por zero n√£o √© permitida"
        return a / b
    else:
        return "Erro: Opera√ß√£o inv√°lida. Use: 'soma', 'subtracao', 'multiplicacao' ou 'divisao'"


try:
    num1 = float(input("Primeiro n√∫mero: "))
    num2 = float(input("Segundo n√∫mero: "))
    oper = input("Opera√ß√£o (soma/subtracao/multiplicacao/divisao): ")

    resultado = calculadora(num1, num2, oper)
    print(f"\nResultado: {resultado}")

except ValueError:
    print("Erro: Digite n√∫meros v√°lidos!")
```

 Lembre-se de usar operadores l√≥gicos como and para definir os intervalos de notas de forma clara.
